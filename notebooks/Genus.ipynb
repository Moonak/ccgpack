```python
%matplotlib inline
import numpy as np 
import pylab as plt
import os
import ccgpack as ccg
```

```python
trshs = np.linspace(-4,4,100)
for i in range(5):
    data = np.load('sample_'+str(i)+'.npy')
    data = data-data.mean()
    data = data/data.std()
    genus = ccg.genus(data,trshs)
    plt.plot(trshs,genus)
    
    
```
